```python
'''This class was created only for test with DC voltage =180V
    * The channel 1 must be the tip voltage with tension
    * The channel 2 must be the resistance voltage to measure the current
    * The channel 3 must be the pickle voltage
    * The channel 4 must be the mains or the photocell voltage
    
    Note: Measurements were taken directly without resistive splitters, parameters
    such as resistance to measure current or filter capacitance for DC can vary.
'''

#White resistor = 2.26 ohm           1ohm resistor = 1.227 ohm
class Pickles_DC:  
    def __init__(self,n_capacitor,n_current,n_pickle,n_mains,r=1.227,columns = ["t","v"]):
        self.t = pd.read_csv("TEK000{}.CSV".format(n_capacitor), sep= ",", decimal=".", names = columns, skiprows = 16)["t"]
        self.Capacitor = pd.read_csv("TEK000{}.CSV".format(n_capacitor), sep= ",", decimal=".", names = columns, skiprows = 16)["v"] #Resistance
        self.Current = pd.read_csv("TEK000{}.CSV".format(n_current), sep= ",", decimal=".", names = columns, skiprows = 16)["v"]/r #Photocell
        self.Pickle = pd.read_csv("TEK000{}.CSV".format(n_pickle), sep= ",", decimal=".", names = columns, skiprows = 16)["v"] #Pickle
        self.Mains = pd.read_csv("TEK000{}.CSV".format(n_mains), sep= ",", decimal=".", names = columns, skiprows = 16)["v"] #Mains    
        self.fs = 1/float(pd.read_csv("TEK000{}.CSV".format(n_capacitor), sep= ",", decimal=".", names = columns)["v"][5])
        self.comment = ""
    
    def Comment(self, texComment):
        self.comment = texComment
    
```

<h4> Measurements notes </h4>

The current measurements were made with a 2.26 ohm resistor and a capacitor of $220\mu f$

The DC filter was made with a capacitor of $1020\mu f$

At least 100V were needed to generate spark.

<ul>
<li>  Test 24 to 27 <strong>pickle didn't glow</strong>. Fist conduction was identified and the pickle was heated. Then the pickle voltage dropped to 0, which indicates that it was positively isolated, and current dropped to 0 with negative peaks. </li>
<li> 28 al 31  <strong>pickle glowed</strong>.The pickle voltage was mainly positive, which indicates that it was negatively isolated and therefore arc was generated.  </li>  
    
</ul>

Technical note: Possible reason for measurement with negative current peaks is measurement noise due to power resistance


```python
Test24_27 = Pickles_DC("24","25","26","27",2.26)
```

```python
Test28_31 = Pickles_DC("28","29","30","31",2.26)
```

```python
fig = plt.figure(figsize=(14, 8))

plt.subplot(211)
plt.ylabel('Voltage [V]', fontsize = 15)
plt.plot(Test24_27.t, Test24_27.Capacitor, marker= '',linestyle = '-', label = 'Capacitor')
plt.plot(Test24_27.t, Test24_27.Pickle, marker= '',linestyle = '-', label = 'Pickle')

plt.legend(fontsize = 12, loc=0)
plt.grid()

#  ----------------------------------------------------------------------------------   #

plt.subplot(212)
plt.ylabel('Current [A]', fontsize = 15)
plt.xlabel('Time [s]', fontsize = 15)
plt.plot(Test24_27.t, Test24_27.Current, marker= '',color="g",linestyle = '-', label = 'Pickle')
plt.legend(fontsize = 12, loc=0)
plt.grid()
```
