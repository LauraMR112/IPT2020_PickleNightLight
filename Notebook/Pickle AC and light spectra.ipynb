# Data 28/01/20 : little pickle

```python
columns = ['t', 'v']
Ch3pic2 = pd.read_csv("Data/28_01_20/TEK00003.CSV", sep= ",", decimal=".", names = columns, skiprows = 16) #Resistance voltage
Ch2pic2 = pd.read_csv("Data/28_01_20/TEK00004.CSV", sep= ",", decimal=".", names = columns, skiprows = 16) #Photocell voltage
Ch1pic2 = pd.read_csv("Data/28_01_20/TEK00005.CSV", sep= ",", decimal=".", names = columns, skiprows = 16) #Mains voltage
```

```python
fig = plt.figure(figsize=(14, 6))
plt.ylabel('Voltage [V]', fontsize = 15, )
plt.xlabel('Time [s]', fontsize = 15)

plt.plot(Ch2pic2['t'], Ch1pic2['v'], marker= 'o',color = 'green',linestyle = '-', label = 'AC voltage (wall socket)')
plt.plot(Ch2pic2['t'], Ch3pic2['v']*10, marker= 'o',color = 'blue',linestyle = '-', label = 'R 5.1 voltage (pickle current)')
# Pickle voltage was increased x10 for better visualization
plt.plot(Ch2pic2['t'],Ch2pic2['v'], marker= 'o',linestyle = '-', label = 'Photoresistor')

plt.legend(fontsize = 12, loc=0)
plt.grid()
```

The objective here is to fit a sine function with the same frequency as the mains (60 Hz) to check if the pickle voltage has the same frequency. It was necessary also to add an amplitude increasing along the time to fit the data properly.

```python
def pickle_voltage(r, a, b, c):
    return (a+b*r)*np.sin(120*np.pi*r + c)

popt1, pcov1 = curve_fit(pickle_voltage, Ch3pic2['t'][8786:9450], Ch3pic2['v'][8786:9450], [9.5, 2.1 , 0.5])
print(popt1)

fig = plt.figure(figsize=(14, 6))
plt.ylabel('Voltage [V]', fontsize = 15, )
plt.xlabel('Time [s]', fontsize = 15)

fit2 = [pickle_voltage(i, popt1[0], popt1[1], popt1[2]) for i in np.linspace(Ch3pic2['t'][8786], Ch3pic2['t'][9450], 10000)]
plt.plot(np.linspace(Ch3pic2['t'][8790], Ch3pic2['t'][9450], 10000), fit2, marker= 'o',color = 'm',linestyle = '-', label = 'Fit')

plt.plot(Ch2pic2['t'], Ch3pic2['v'], marker= 'o',color = 'blue',linestyle = '-', label = 'Pickle')

plt.xlim((-1,-0.4))
plt.legend(fontsize = 10, loc=0)
plt.grid()
```
